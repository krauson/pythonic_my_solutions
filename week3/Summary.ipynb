<p style="text-align: right; direction: rtl; float: right; clear: both;">
פונקציית הגיבוב שלנו תעבוד כך:<br>
</p>
<ol style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>נאתחל משתנה עזר בשם <var>hash</var> כך שערכו יהיה 1.</li>
    <li>עבור כל אות במחרוזת:</li>
    <ol>
        <li>נחשב את <code dir="ltr" style="direction: ltr;">ord(letter)</code> ונכפיל במשתנה העזר <var>hash</var>.</li>
        <li>נכפיל את הערך שקיבלנו במיקום של האות הבאה במחרוזת.
        <li>לתוצאה הזו נבצע מודולו 397,643, ונשמור על <var>hash</var>.</li>
    </ol>
    <li>כדי שהפלט תמיד יהיה באורך זהה, נשתמש במודולו 100,297 על התוצאה (חשבו: איך מודולו גורם לזה לקרות?)</li>
</ol>
<p style="text-align: right; direction: rtl; float: right; clear: both;">
פונקציית הגיבוב שיצרנו מחזירה תמיד ערכים באורך קבוע (בין 0 ל־100,297), כפי שפונקציית גיבוב צריכה להחזיר (לאו דווקא באורך זה, אבל הפלט חייב להיות באורך קבוע).
</p>

```python
#upload 49

def gibuv(param_str):


    temp_hash = 1
    i = 0
    
    while i < len(param_str):

        j = i + 1
        temp_hash = (temp_hash * ord(param_str[i]) * j) % 397643
        i += 1

    return temp_hash % 100297


print(gibuv('aba'))
print(gibuv('i calculate hashes for fun'))
print(gibuv('10'))
```

```python
#upload 50
def print_board(board):
    
    temp_board = board.copy()
    
    i = 0
    while i < len(temp_board):
    
        temp_board[i] = "  ".join(temp_board[i])
        i += 1
    
    temp_board = "\n".join(temp_board)
    
    print(temp_board)
    
    return

def is_cell_empty(places , board):


    if board[places[0]][places[1]] != '-':
        print(f"Cell ({places[0]}, {places[1]}) is taken, use other")
        print_board(board)
        return False

    else:
        return True


def places_are_valid(places):


    line_num = places[0]
    column_num = places[1]

    # places are between 0 - 2

    if (0 > line_num) or 2 < line_num:
        print(f"Invalid line chosen ({line_num}).")
        return False

    if (0 > column_num) or 2 < column_num:
        print(f"Invalid column chosen ({column_num}).")
        print(places)
        return False

    return True


def convert_to_int_list(param_str):


    num_list = param_str.split()

    i = 0
    while i < len(num_list):

        num_list[i] = int(num_list[i])
        i += 1

    return num_list

def make_turn(player_char,board):

    places = input(f"Player '{player_char}' Please choose a cell:")

#   convert to list of ints
    places = convert_to_int_list(places)

    print(places)

    if places_are_valid(places) and is_cell_empty(places , board):

        board[places[0]][places[1]] = player_char
        print_board(board)
        return  board

    else:
        return make_turn(player_char, board)


def check_board(board):


        i = 0
        while i < len(board):

            if board[i][0] == board[i][1] == board[i][2] and board[i][0] != '-':
                return board[i][0]
            i += 1

        # checking if there is win in one of the columns

        i = 0
        while i < len(board):

            if board[0][i] == board[1][i] == board[2][i] and board[0][i] != '-':
                return board[0][i]
            i += 1

        # checking if there is win in one of the crosses

        if board[0][0] == board[1][1] == board[2][2] and board[1][1] != '-':
            return board[0][0]

        elif board[0][2] == board[1][1] == board[2][0] and board[1][1] != '-':
            return board[0][2]

        else:
            return ''


def tic_tac_toe():


    moves_counter = 0

    board = [['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']]

    player_char = 'O'
    
    print_board(board)

    while check_board(board) == '' and moves_counter < 9:

        if moves_counter % 2 == 0:

            player_char = 'O'

        else:

            player_char = 'X'

        board = make_turn(player_char, board)

        moves_counter += 1


    if check_board(board) == '':
        print("It's a tie!")

    else:
        print(f"The winner is: {player_char}")

    answer = input("press y if you want to play another game")

    if answer == "y":
        tic_tac_toe()

    else:
        print("Goodbye! I hope you enjoyed:)")
        return


tic_tac_toe()
```

```python
## upload 51

def gibuv(param_str):


    temp_hash = 1
    i = 0
    while i < len(param_str):

        j = i + 1
        temp_hash = (temp_hash * ord(param_str[i]) * j) % 397643
        i += 1
    return temp_hash % 100297


def remove_user(username , passwd):


    print("Im in remove_user func")
    with open('resources/passwd.txt' , 'r') as passwd_file:

        # receive passwords as list of str

        passwd_list = passwd_file.readlines()

        # create a list of lists
        i = 0
        while i < len(passwd_list) :

            passwd_list[i] = passwd_list[i].split(":")
            i += 1

        # finding the inner list with the username to remove
        i = 0
        while i < len(passwd_list):

            if passwd_list[i][0] == username:
                place_of_username = i
            i += 1
        # returning the ':' to the list

        j = 0
        while j < len(passwd_list):

            passwd_list[j] = ":".join(passwd_list[j])
            j += 1

        # slicing out the username:
        new_passwd_list= passwd_list[:place_of_username] + passwd_list[place_of_username + 1:]

        # convert to  str:
        new_passwd_list = "".join(new_passwd_list)

        #  using write mode to write the new str:
        with open('resources/passwd.txt' , 'w') as passwd_file:

            passwd_file.write(new_passwd_list)
            return


def authenticate(username , user_passwd): # checks of password is correct


    with open('resources/passwd.txt' , 'r') as passwd_file:
        print("Im in the authenticate func")

        passwd_list = passwd_file.readlines()

        i = 0
        while i < len(passwd_list) :

            passwd_list[i] = passwd_list[i].split(":")

            if passwd_list[i][0] == username: # check username

                if passwd_list[i][1] == user_passwd:  # checks password
                    return True
                else:
                    return False
            i += 1


def check_username(username):


    with open('resources/passwd.txt' , 'r') as passwd_file: 

        print("Im in check username func!")
        passwd_list = passwd_file.readlines()

        i = 0
        while i < len(passwd_list):
            passwd_list[i] = passwd_list[i].split(":")
            if passwd_list[i][0] == username:
                return True

            else:
                i += 1 # return to main


def check_password(username):


    print("Im in check_password func")
    with open('resources/passwd.txt' , 'r') as passwd_file:

        passwd_list = passwd_file.readlines()

        counter = 1
        while counter <= 3:

            # manipulations on passwd to fit to passwsd from list
            passwd = input("Enter the password:")

            passwd = gibuv(passwd)

            passwd_to_send = str(passwd) + "\n"

            if authenticate(username , passwd_to_send):
                return True # to main

            else:
                print(f"Wrong password ({counter}/3).)")
                counter += 1

        if counter > 3:

            remove_user(username , passwd)
            print(f"\'{username}\' was removed.")
            return


def main():


    with open('resources/passwd.txt' , 'r') as passwd_file:

        passwd_list = passwd_file.readlines()

        print(passwd_list)

        username = input("Enter the username:")

        if check_username(username):

            if check_password(username):

                print(f"Welcome \'{username}\'.")
                return

        else:
            print(f"\'{username}\' is not recognized as a registered username.")
            return


def register_to_bank():


    username = input("Enter the username you would like to use:")

    with open('resources/passwd.txt' , 'r') as passwd_file:

        passwd_lines = passwd_file.read()

        if check_username(username):
            print(f"An account with username: \"{username}\" already exist, an account was not created. ")
            return

        # username doesn't exist. so creating an account
        else:        
            passwd = input("Enter password:")          
            hash_num = str(gibuv(passwd))
            item_to_add = "\n" + username + ":" + hash_num

            with open('resources/passwd.txt' , 'a') as passwd_file:

                passwd_file.write(item_to_add)
                print(f"Account with the name: \"{username}\" was created and added to the bank.")
                return


#register_to_bank()   
main()
```
