```python
#upload 67
def get_votes_list():

    with open('resources/elections.txt' , 'r') as elections_file:

        votes_list = elections_file.read()
        votes_list = votes_list.split("\n")
        return votes_list

def get_votes_dict(votes_list):

    # putting all the kind of parties in the dict:

    votes_dict = {}
    party_votes_count = 0

    for party_name in votes_list:

        if party_name not in votes_dict:

            party_votes_count = votes_list.count(party_name)
            votes_dict[party_name] = party_votes_count

    print(votes_dict)
    return votes_dict

def passed_block_precentage(votes_dict):

    passed_dict = {}
    didnt_passed_dict = {}

    for party_name, party_result in votes_dict.items():

        if votes_dict[party_name] >= 35340:
            passed_dict[party_name] = party_result
        else:
            didnt_passed_dict[party_name] = party_result
    print("passed the block precentage:\n")
    print(passed_dict)
    print()
    print("didn't passed the block precentage:\n")       
    print(didnt_passed_dict)

def get_votes_for_passed_parties(votes_dict):

    passed_dict = {}
    didnt_passed_dict = {}
    passed_parties_sum_of_votes = 0

    for party_name, party_result in votes_dict.items():

        if votes_dict[party_name] >= 35340:
            passed_dict[party_name] = party_result
            passed_parties_sum_of_votes += party_result

    print(f"The sum of votes for the passed parties is : {passed_parties_sum_of_votes}.")

    return passed_dict

def get_mandate_dict(votes_for_passed_parties_dict , votes_per_mandate):

    mandate_dict = {}

    for party_name, party_votes in votes_for_passed_parties_dict.items():

        mandate_dict[party_name] = round(party_votes / votes_per_mandate)

    print(mandate_dict)

    return mandate_dict

def get_votes_per_mandate_ratio(votes_per_mandate_dict):

    print("im in votes per mandate ratio")
    sum_of_votes = 0

    for party_name, party_votes in votes_per_mandate_dict.items():

        sum_of_votes += party_votes

    votes_per_mandate = round(sum_of_votes / 120)

    return votes_per_mandate

def print_possible_coalition(mandate_dict):

    MANDATE_FOR_COALITION = 61

    possible_coalition_for_NilsIsALeader = []
    possible_coalition_for_MidgeLandIsUs = []

    NilsIsALeader_mandates = mandate_dict['NilsIsALeader']
    MidgeLandIsUs_mandates = mandate_dict['MidgeLandIsUs']
    
    for party_name , certain_party_mandetes in mandate_dict.items():

    # possible coalition for NilsIsALeader:
       if party_name != 'MidgeLandIsUs' and party_name != 'NilsIsALeader':

            if certain_party_mandetes + NilsIsALeader_mandates >= 61:
                 possible_coalition_for_NilsIsALeader.append(party_name)

            if  certain_party_mandetes + MidgeLandIsUs_mandates >= 61:
                 possible_coalition_for_MidgeLandIsUs.append(party_name)

    print('\n') 
    print(f"possible coalition for NilsIsALeader is:")

    if len(possible_coalition_for_NilsIsALeader) == 0:
        print("There are no possible coalition for possible_coalition for NilsIsALeader.\n")

    else:
        for party_name in possible_coalition_for_NilsIsALeader:

            print(party_name)

    print(f"possible coalition for MidgeLandIsUs is:")

    if len(possible_coalition_for_MidgeLandIsUs) == 0:
        print("There are no possible coalition for possible coalition for NilsIsALeader MidgeLandIsUs.\n")

    else:
        for party_name in possible_coalition_for_MidgeLandIsUs:

            print(party_name)

def get_election_results():

    votes_list = get_votes_list()

    votes_dict = get_votes_dict(votes_list)

    passed_block_precentage_dict = passed_block_precentage(votes_dict)

    votes_for_passed_parties = get_votes_for_passed_parties(votes_dict)

    votes_per_mandate = get_votes_per_mandate_ratio(votes_for_passed_parties)

    print(f"votes per mandate is {votes_per_mandate}.")

    mandate_dict = get_mandate_dict(votes_for_passed_parties ,votes_per_mandate)

    print_possible_coalition(mandate_dict)
get_election_results()
```
